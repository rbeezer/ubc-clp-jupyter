<div class="mathbook-content"><header title="Subsection 3.9.1 Statement of the chain rule"><h3 class="heading hide-type" alt="Subsection 3.9.1 Statement of the chain rule"><span class="type">Subsection</span><span class="codenumber">3.9.1</span><span class="title">Statement of the chain rule</span></h3></header></div>

<div class="mathbook-content"><article class="theorem-like" id="thm_DIFFchainRuleV1"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">3.9.2</span><span class="title">The chain rule  — version 1</span></h6><p id="p-1819">Let $a \in \mathbb{R}$ and let $g(x)$ be a function that is differentiable at $x=a\text{.}$ Now let $f(u)$ be a function that is differentiable at $u=g(a)\text{.}$ Then the function $F(x) = f(g(x))$ is differentiable at $x=a$ and</p>
\begin{align*}
F'(a) &=f'\big(g(a)\big)\,g'(a)
\end{align*}
</article></div>

<div class="mathbook-content"><p id="p-1820">Here, as was the case earlier in this chapter,  we have been very careful to give the point at which the derivative is evaluated a special name (i.e. $a$). But of course this evaluation point can really be any point (where the derivative is defined). So it is very common to just call the evaluation point “$x$” rather than give it a special name like “$a$”, like this:</p></div>

<div class="mathbook-content"><article class="theorem-like" id="thm_DIFFchainRuleV2"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">3.9.3</span><span class="title">The chain rule — version 2</span></h6><p id="p-1821">Let $f$ and $g$ be differentiable functions then</p>
\begin{align*}
\diff{}{x} f\big( g( x) \big) &= f'\big(g(x)\big) \cdot g'(x)
\end{align*}
</article></div>

<div class="mathbook-content"><p id="p-1822">Notice that when we form the composition $f\big(g(x)\big)$ there is an “outside” function (namely $f(x)$) and an “inside” function (namely $g(x)$). The chain rule tells us that when we differentiate a composition that we have to differentiate the outside and then multiply by the derivative of the inside.</p>
\begin{align*}
\diff{}{x} f\big( g( x) \big) 
&= \underbrace{f'\big(g(x)\big)}_\text{diff outside} \cdot 
\underbrace{g'(x)}_\text{diff inside}
\end{align*}
<p>Here is another statement of the chain rule which makes this idea more explicit.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="thm_DIFFchainRuleV3"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">3.9.4</span><span class="title">The chain rule — version 3</span></h6><p id="p-1823">Let $y = f(u)$ and $u = g(x)$ be differentiable functions, then</p>
\begin{align*}
\diff{y}{x} &= \diff{y}{u} \cdot \diff{u}{x}
\end{align*}
</article></div>

<div class="mathbook-content"><p id="p-1824">This particular form is easy to remember because it looks like we can just “cancel” the $\mathrm{d}u$ between the two terms.</p>
\begin{align*}
\diff{y}{x} &= \frac{\mathrm{d}{y}}{\cancel{\mathrm{d}u}} \cdot 
\frac{\cancel{\mathrm{d}{u}}}{\mathrm{d}x} 
\end{align*}
</div>

<div class="mathbook-content"><p id="p-1825">Of course,  $\mathrm{d}u$ is not, by itself, a number or variable <a knowl="" class="id-ref" refid="hk-fn-82" id="fn-82"><span class="footnote"><sup> 1 </sup></span></a><span id="hk-fn-82" class="hidden-content tex2jax_ignore"><span class="footnote">In this context $\mathrm{d}u$ is called a differential. There are ways to understand and manipulate these in calculus but they are beyond the scope of this course.</span></span> that can be cancelled. But this is still a good memory aid.</p></div>

<div class="mathbook-content"><p id="p-1826">The hardest part about applying the chain rule is recognising when the function you are trying to differentiate is really the composition of two simpler functions. This takes a little practice. We can warm up with a couple of simple examples.</p></div>

<div class="mathbook-content"><article class="example-like" id="eg_DIFFchainruleA"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.9.5</span><span class="title">Derivative of a power of $\sin x$</span></h6><p id="p-1827">Let $f(u) = u^5$ and $g(x) = \sin(x)\text{.}$ Then set $F(x) = f\big(g(x)\big) = \big(\sin(x)\big)^5\text{.}$ To find the derivative of $F(x)$ we can simply apply the chain rule  — the pieces of the composition have been laid out for us. Here they are.</p>
\begin{align*}
f(u) &= u^5 & f'(u) &= 5u^4 \\
g(x) &= \sin(x) & g'(x) &= \cos x \\

\end{align*}
<p>We now just put them together as the chain rule tells us</p>
\begin{align*}
\diff{F}{x} &= f'\big(g(x)\big) \cdot g'(x) \\
&= 5\big(g(x)\big)^4 \cdot \cos(x) & \text{since } f'(u) = 5u^4 \\
&= 5 \big(\sin(x)\big)^4 \cdot \cos(x) 
\end{align*}
<p id="p-1828">Notice that it is quite easy to extend this to any power. Set $f(u) = u^n\text{.}$ Then follow the same steps and we arrive at</p>
\begin{align*}
F(x) &= (\sin(x))^n & F'(x) &= n \big(\sin(x) \big)^{n-1} \cos(x)
\end{align*}
</article></div>

<div class="mathbook-content"><p id="p-1829">This example shows one of the ways that the chain rule appears very frequently — when we need to differentiate the power of some simpler function. More generally we have the following.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-77"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.9.6</span><span class="title">Derivative of a power of a function</span></h6><p id="p-1830">Let $f(u) = u^n$ and let $g(x)$ be any differentiable function. Set $F(x) = f\big(g(x)\big) = g(x)^n\text{.}$ Then</p>
\begin{align*}
\diff{F}{x} = \diff{}{x} \big( g(x)^n \big) &= n g(x)^{n-1} \cdot g'(x)
\end{align*}
<p>This is precisely the result in Example <a href="subsection-22.ipynb#eg_DIFFsimpleToolsA" alt="Example 3.6.6 Extending the product rule to more than two factors" title="Example 3.6.6 Extending the product rule to more than two factors">3.6.6</a>  and Lemma <a href="subsection-22.ipynb#lem_diff_ftothen" alt="Lemma 3.6.7 " title="Lemma 3.6.7 ">3.6.7</a>.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-78"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.9.7</span><span class="title">Derivative of $\cos(3x-2)$</span></h6><p id="p-1831">Let $f(u) = \cos(u)$ and $g(x) = 3x-2\text{.}$ Find the derivative of</p>
\begin{align*}
F(x) &= f\big(g(x)\big) = \cos(3x-2).
\end{align*}
<p id="p-1832">Again we should approach this by first writing down $f$ and $g$ and their derivatives and then putting everything together as the chain rule tells us.</p>
\begin{align*}
f(u) &= \cos(u) & f'(u) &= -\sin(u) \\
g(x) &= 3x-2 & g'(x) &= 3 
\end{align*}
<p>So the chain rule says</p>
\begin{align*}
F'(x) &= f'\big(g(x)\big) \cdot g'(x) \\
&= -\sin\big( g(x) \big) \cdot 3 \\
&= -3 \sin(3x-2)
\end{align*}
</article></div>

<div class="mathbook-content"><p id="p-1833">This example shows a second way that the chain rule appears very frequently — when we need to differentiate some function of $ax+b\text{.}$ More generally we have the following.</p></div>

<div class="mathbook-content"><article class="example-like" id="eg_DIFFchainaxb"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.9.8</span><span class="title">Derivative of $f(ax+b)$</span></h6><p id="p-1834">Let $a,b \in \mathbb{R}$ and let $f(x)$ be a differentiable function. Set  $g(x) = ax+b\text{.}$ Then</p>
\begin{align*}
\diff{}{x} f(ax+b) &= \diff{}{x} f\big(g(x)\big) \\
&= f'\big(g(x)\big) \cdot g'(x) \\
&= f'(ax+b) \cdot a
\end{align*}
<p>So the derivative of $f(ax+b)$ with respect to $x$ is just $a f'(ax+b)\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1835">The above is a very useful result that follows from the chain rule, so lets make it a corollary to highlight it.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="cor_f_of_axb"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">3.9.9</span></h6><p id="p-1836">Let $a,b \in \mathbb{R}$ and let $f(x)$ be a differentiable function, then</p>
\begin{align*}
\diff{}{x} f(ax+b) &= af'(ax+b).
\end{align*}
</article></div>

<div class="mathbook-content"><article class="example-like" id="eg_DIFFcampfireB"><h6 class="heading"><span class="type">Example</span><span class="codenumber">3.9.10</span><span class="title"><a href="sec_chain_rule.ipynb#eg_DIFFcampfire" alt="Example 3.9.1 Walking towards a campfire" title="Example 3.9.1 Walking towards a campfire">3.9.1</a> continued</span></h6><p id="p-1837">Let us now go back to our motivating campfire example. There we had</p>
\begin{align*}
f(x) &= \text{ temperature at position $x$} \\
x(t) &= \text{ position at time $t$} \\
F(t) &= f(x(t)) = \text{ temperature at time $t$}
\end{align*}
<p>The chain rule gave</p>
\begin{align*}
F'(t) &= f'\big(x(t)\big) \cdot x'(t)
\end{align*}
<p>Notice that the units of measurement on both sides of the equation agree — as indeed they must. To see this, let us assume that $t$ is measured in seconds, that $x(t)$ is measured in metres and that $f(x)$ is measured in degrees. Because of this $F(x(t))$ must also be measured in degrees (since it is a temperature).</p><p id="p-1838">What about the derivatives? These are rates of change. So </p><ul id="ul-129" style="list-style-type: disc;"><li id="li-496">$F'(t)$ has units $\frac{\rm degrees}{\rm second}\text{,}$</li><li id="li-497">$f'(x)$ has units $\frac{\rm degrees}{\rm metre}\text{,}$ and</li><li id="li-498">$x'(t)$ has units $\frac{\rm metre}{\rm second}\text{.}$</li></ul><p> Hence the product</p>
\begin{align*}
f'\big(x(t)\big) \cdot x'(t) & 
\text{ has units } = \frac{\rm degrees}{\rm metre} \cdot 
\frac{\rm metre}{\rm second} 
= \frac{\rm degrees}{\rm second}.
\end{align*}
<p>has the same units as $F'(t)\text{.}$ So the units on both sides of the equation agree. Checking that the units on both sides of an equation agree is a good check of consistency, but of course it does not prove that both sides are in fact the same.</p></article></div>
